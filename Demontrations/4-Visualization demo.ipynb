```python
val = 3+10*rand(5)    # the bar lengths
pos = arange(5)+.5    # the bar centers on the y axis

figure(1)
barh(pos,val, align='center')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Accuracy')
title('Performance of methods')
grid(True)
```

```python
figure(2)
barh(pos,val, xerr=rand(5), ecolor='r', align='center')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Performance')
```

```python
val1 = 3+10*rand(5)    # the bar lengths
val2 = 2+10*rand(5)    # the bar lengths
pos = arange(5)+.5    # the bar centers on the y axis
h=0.3

figure(figsize=[10,5])
before = barh(pos,val1, height=h, color='b')
after = barh(pos+h, val2, height=h, color='r')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Accuracy')
title('Performance of methods')

legend( (before, after), ('before transformation', 'after transformation') )

grid(True)
```

## Stacked bars

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, color='r', yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, color='y',
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind + width/2., ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
```

## Displaying 2D functions

```python
def fun(x, y):
    return x**2+(y-3)**2
```

```python
n = 256
X = np.linspace(-6, 6, n)
Y = np.linspace(-6, 6, n)
#XX, YY = np.meshgrid(x, y)


F = zeros([n,n])
for (ix,x) in enumerate(X):
    for (iy,y) in enumerate(Y):
        F[ix,iy] = fun(x,y)
        
plt.contourf(X, Y, F.T, 18, c='black', alpha=.75, cmap='hot')
colorbar()
C = plt.contour(X, Y, F.T, 18, colors='black', linewidth=.5)
xlabel('x')
ylabel('y')
```

```python
def fun(x, y):
    return (x-1)**2+y**2

n = 10
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)
plt.imshow(fun(X, Y))      # but in images grid structure should be regular
colorbar()
```

# 3D plots

```python
mpld3.disable_notebook()  # important (otherwise will not work)
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2+Y**2

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
