```python
def fun(x, y):
    return (x-1)**2+y**2

n = 256
X = np.linspace(-6, 6, n)
Y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)


F = zeros([n,n])
for (ix,x) in enumerate(X):
    for (iy,y) in enumerate(Y):
        F[ix,iy] = fun(x,y)
        
plt.contourf(X, Y, F.T, 8, alpha=.75, cmap='hot')
colorbar()
C = plt.contour(X, Y, F.T, 8, colors='black', linewidth=.5)
xlabel('x')
ylabel('y')
```

# 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2+Y**2

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
